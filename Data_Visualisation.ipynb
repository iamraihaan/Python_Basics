5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Set random seed for reproducibility
np.random.seed(42)

# Step 2: Generate synthetic data with correlated features
# For this example, let's assume a dataset with 5 features
mean = [0, 0, 0, 0, 0]  # Mean of the distribution for each feature
covariance_matrix = [
    [1, 0.8, 0.5, 0.3, 0.1],  # Feature 1 correlations
    [0.8, 1, 0.6, 0.4, 0.2],  # Feature 2 correlations
    [0.5, 0.6, 1, 0.5, 0.3],  # Feature 3 correlations
    [0.3, 0.4, 0.5, 1, 0.6],  # Feature 4 correlations
    [0.1, 0.2, 0.3, 0.6, 1]   # Feature 5 correlations
]
data = np.random.multivariate_normal(mean, covariance_matrix, size=1000)
df = pd.DataFrame(data, columns=['Feature1', 'Feature2', 'Feature3', 'Feature4', 'Feature5'])

# Step 3: Calculate correlation matrix
correlation_matrix = df.corr()

# Step 4: Plot correlation matrix using Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)
fig = px.scatter_3d(df, x='X', y='Y', z='Z')
fig.show()
```

```python
#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

from bokeh.plotting import figure, show
from bokeh.io import output_notebook


x = np.linspace(0, 10, 100)
y = np.sin(x)
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='y')
p.scatter(x, y, legend_label="sine", line_width=2)
show(p)
```
